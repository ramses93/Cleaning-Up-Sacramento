```python
import pandas as pd
```

```python
rawdata_2008 = pd.read_csv('Data/data.csv')
```

```python
rawdata_2008
```

```python
cleandata_2008 = rawdata_2008.copy()
cleandata_2008 = cleandata_2008.drop(cleandata_2008.index[cleandata_2008['sq_ft']==0])
display(cleandata_2008.info())
display(cleandata_2008.describe)
```

```python
stats_2008 = cleandata_2008.copy()
stats_2008['price per sqft'] = stats_2008['price']/stats_2008['sq_ft']
stats_2008
```

```python
zip_stats = (round(
    stats_2008.copy()
    .groupby('zip')
    .agg({'price':'median','price per sqft':'median','sq_ft':'median','city':'count'})
    .rename(columns={'city':"Sales",'price':"Median Price",'price per sqft':'Median Price/sqft','sq_ft':"Median sqft"})
    ,0))
```

```python
zip_stats
```

```python
zip_stats.to_csv('Data/zipcode_medians_may2008.csv')
```
