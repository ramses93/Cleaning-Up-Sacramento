```python
import pandas as pd
```

```python
rawdata_2008 = pd.read_csv('Data/data.csv')
rawdata_2018 =pd.read_csv('Data/zipcode_sales_nov2018.csv')
```

```python
display(rawdata_2008)
display(rawdata_2018)
```

```python
cleandata_2018= (rawdata_2018.copy()
                 .rename(columns={"ZIP":'zip',
                                  "Sales":'Sales_2018',
                                  '2017 median':'Median_Price_2017',
                                  '2018 median':'Median_Price_2018'})
                 .drop(['Change'],axis=1))
display(cleandata_2018)
```

```python
cleandata_2008 = rawdata_2008.copy()
cleandata_2008 = cleandata_2008.drop(cleandata_2008.index[cleandata_2008['sq_ft']==0])
display(cleandata_2008.info())
display(cleandata_2008.describe)
```

```python
stats_2008 = cleandata_2008.copy()
stats_2008['price per sqft'] = stats_2008['price']/stats_2008['sq_ft']
stats_2008
```

```python
zip_stats = (round(
    stats_2008.copy()
    .groupby('zip')
    .agg({'price':'median',
          'price per sqft':'median',
          'sq_ft':'median',
          'city':'count'})
    .rename(columns={'city':"Sales_2008",
                     'price':"Median_Price_2008",
                     'price per sqft':'Median_Price_sqft_2008',
                     'sq_ft':"Median_sqft_2008"})
    ,0))
```

```python
zip_stats
```

```python
zip_stats.to_csv('Data/zipcode_sales_may2008.csv')
```

```python
combined_stats = pd.merge(zip_stats,cleandata_2018, on="zip",how='inner')
```

```python
display(combined_stats)
```

```python
combined_stats.to_csv('Data/zipcode_medians_combined.csv')
```
